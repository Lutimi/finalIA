```python
import numpy as np 
import pandas as pd
from perceptron import *
```

```python
df = pd.read_csv("iris.data", header = None)
df.head()
```

```python
X = df.iloc[ :100, [0,2]].values
D = df.iloc[ :100, 4].values
D = np.where(D == "Iris-setosa", -1, 1)
net = PreceptronSimple()
```

```python
net.train(X, D, 6)
print(net.W)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X[:50, 0], X[:50, 1], color = "blue")
plt.scatter(X[50:100, 0], X[50:100, 1], color = "red")
#hiperplano
m, b = -1*(net.W[1]/net.W[2]),  -1*(net.W[0]/net.W[2])  # slope and intercept 
x = np.linspace(3,8,100)
plt.plot(x, x*m + b)
```

```python
Xval = [5, 5]
target = net.predict(Xval)
print(Xval, target)
```
